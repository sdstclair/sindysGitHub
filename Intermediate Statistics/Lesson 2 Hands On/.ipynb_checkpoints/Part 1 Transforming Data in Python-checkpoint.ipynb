# Number of trips Winter

```python
seattlePR['# of trips Winter'].hist()
```

```python
sns.distplot(seattlePR['# of trips Winter'])
```

## This is more positively skewed than not. But lacks an even bell shaped distribution. This would be considered bimodial.

```python
seattlePR['# of trips WinterSQRT'] = np.sqrt(seattlePR['# of trips Winter'])
```

```python
seattlePR['# of trips WinterSQRT'].hist()
```

```python
sns.distplot(seattlePR['# of trips WinterSQRT'])
```

# Still bimodial. So this would cause me to continue in transforming this data using log.

```python
seattlePR['# of trips WinterLOG'] = np.log(seattlePR['# of trips Winter'])
```

```python
seattlePR['# of trips WinterLOG'].hist()
```

# This appears to be more distributed, in a more negative sense anyway.

```python
sns.distplot(seattlePR['# of trips WinterLOG'])
```

# But in this plot of the histogram, it appears to be a bimodial distribution that is more towards the normal/negative side.
