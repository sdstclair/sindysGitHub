## Lesson 3 Recoding Continuous Variable Column into String Column

```python
def age_recode (series): 
    if series < 18:
        return "minor"
    if series > 18: 
        return "adult"
    if series > 55: 
        return "senior"
 
Beauty_LinkedIn['ageR'] = Beauty_LinkedIn['age'].apply(age_recode)
```

```python
Beauty_LinkedIn.ageR
```

```python
Beauty_LinkedIn.head()
```

```python
Beauty_LinkedIn.drop(['age'], axis=1, inplace = True)
```

## Lesson 4: (a) Choose a continuous variable column and create a histogram. (b) Then choose a categorical variable column and create a bar chart.
