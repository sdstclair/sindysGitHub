```r
# Subsetting category dfs to only keep necessary columns
Cat1 <- subset(Cat1, select = c('Commodity_Description', 'Year', 'Value'))
Cat2 <- subset(Cat2, select = c('Commodity_Description', 'Year', 'Value'))
Cat3 <- subset(Cat3, select = c('Commodity_Description', 'Year', 'Value'))
Cat4 <- subset(Cat4, select = c('Commodity_Description', 'Year', 'Value'))
Cat5 <- subset(Cat5, select = c('Commodity_Description', 'Year', 'Value'))
unique(Cat1$Commodity_Description)
unique(Cat2$Commodity_Description)
unique(Cat3$Commodity_Description)
unique(Cat4$Commodity_Description)
unique(Cat5$Commodity_Description)
```

```r
# Appending Category dfs to eachother
CommodUS98_12 <- rbind(Cat1, Cat2, Cat3, Cat4, Cat5)
CommodUS98_12
```

```r
# Dropping multiple values
# Dropping repeat entries for Cat 5
comUS98_12 = aggregate(CommodUS98_12$Value, by=list(CommodUS98_12$Commodity_Description, CommodUS98_12$Year), FUN=max)
comUS98_12
```

```r
# Renaming Columns
comUS98_12 <- rename(comUS98_12, Commodity = Group.1)
comUS98_12 <- rename(comUS98_12, Year = Group.2)
comUS98_12 <- rename(comUS98_12, Metric.Tons = x)
```
